```python
from math import ceil
import numpy as np
from tqdm import tqdm

from tenso.prototypes.heom import system_multibath # Propagator
from tenso.prototypes.bath import gen_bcf # Bath correlation function generator
```

```python
out_name = 'capstone2'
```

```python
bath_simulation1 = gen_bcf(
#    re_d=[540],    # Reorganization energy in cm-1
#    width_d=[70],  # Width of the DL spectral density in cm-1
    freq_b = [1243], # Frequency of the brownian oscillator in cm-1
    re_b = [161.6],   # Reorganization energy of the brownian oscillator in cm-1
    width_b = [10], # Width of the brownian oscillator in cm-1
    temperature=300, # Temperature in Kelvin
    decomposition_method='Pade', # Decomposition method for the bath correlation function
    n_ltc=1, # Number of low-temperature correction terms
)

bath_simulation2 = gen_bcf(
#    re_d=[540],    # Reorganization energy in cm-1
#    width_d=[70],  # Width of the DL spectral density in cm-1
    freq_b = [1243], # Frequency of the brownian oscillator in cm-1
    re_b = [161.6],   # Reorganization energy of the brownian oscillator in cm-1
    width_b = [10], # Width of the brownian oscillator in cm-1
    temperature=300, # Temperature in Kelvin
    decomposition_method='Pade', # Decomposition method for the bath correlation function
    n_ltc=1, # Number of low-temperature correction terms
)
```

```python
# Initialize system Hamiltonian
H = np.array([[0, 0, 0], 
              [0, 1240, 0],
              [0, 0, 0]], dtype=np.complex128)
```

```python
end_time = 100.0 # End time in fs
dt = 1
wfn = np.array([1.0, 0.0, 0.0], dtype=np.complex128)
# Run a single calculation with the default settings:
# frame_method='tree2', stepwise_method='mix', ps_method='vmf'
propagator = system_multibath(
    fname=out_name,
    init_rdo=np.outer(wfn, wfn.conj()),
    sys_ham=H,
    sys_ops=[np.array([[0, 0, 0],
                       [0, 0, 1],
                       [0, 1, 0]],dtype=np.complex128), 
             np.array([[0, 1, 0],
                       [1, 0, 0],
                       [0, 0, 0]],dtype=np.complex128)],
    bath_correlations=[bath_simulation1, bath_simulation2],
    dim=20,
    end_time=end_time,
    stepwise_method = 'simple',
    ps_methods='ps1',
    rank=20,
    step_time=dt,)

progress_bar = tqdm(propagator, total=ceil(end_time / dt))

for _t in (progress_bar):
    progress_bar.set_description(f'@{_t:.2f} fs')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

file_sb = Path("capstone2.dat.log")
time_units = "fs"
pop_labels = [r"$\rho_{00}$", r"$\rho_{11}$"]

plt.rcParams.update({
    "text.usetex": False,
    "font.family": "serif", "font.serif": ["Times"], "font.size": 9,
    "axes.labelsize": 9, "axes.titlesize": 9, "axes.linewidth": 0.8,
    "lines.linewidth": 1.2, "lines.markersize": 4,
    "xtick.direction": "in", "ytick.direction": "in",
    "xtick.top": True, "ytick.right": True,
    "figure.dpi": 300,
})

def load_spinboson_pops(fname, convert_to_ps=False):
    arr = np.genfromtxt(fname, dtype=complex, comments="#", skip_header=1)
    t = arr[:,0].real
    if convert_to_ps:
        t *= 1
    pops = np.vstack((arr[:,1].real, arr[:,4].real)).T
    # purity Tr(rho^2) = rho00^2 + rho11^2 + 2*rho01*rho10
    purity = (arr[:,1]**2 + arr[:,4]**2 + 2 * arr[:,2] * arr[:,3]).real
    return t, pops, purity

# load data
convert = (time_units.lower() == "ps")
t_sb, p_sb, purity_sb = load_spinboson_pops(file_sb, convert_to_ps=convert)

# make figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(6.8, 2.0), sharex=True)

ax1.plot(t_sb, p_sb[:,0], label=pop_labels[0])
ax1.plot(t_sb, p_sb[:,1], label=pop_labels[1])

ax1.set_ylabel("Population", labelpad=6)
ax1.set_xlabel("Time (fs)", labelpad=3)
#ax1.set_xlim(0, 49)
ax1.legend(loc="best", fontsize=8, frameon=False)

ax2.plot(t_sb, purity_sb, color="k")

ax2.set_ylabel(r"Purity $\mathrm{Tr}(\rho^2)$", labelpad=6)
ax2.set_xlabel("Time (fs)", labelpad=3)
#ax2.set_ylim(-10, 10)

fig.tight_layout()
plt.savefig("SpinBoson_simple.png", bbox_inches='tight', dpi=300, transparent=True)
plt.show()
```
